## 3.

2 очка может выпасть на 2 костях только при одном случае: если и в первом, и во втором случае выпало 1. Вероятность выпадения 1 в каждом броске $\frac{1}{6}$, два броска независимы друг от друга, следовательно, вероятность получить в сумме 2 очка равна $\frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36}$.

5 очков может выпасть на 2 костях при следующих бросках: 1 и 4, 2 и 3, 3 и 2, 4 и 1. Следовательно, итоговая вероятность выпадения 5 очков $\frac{4}{36} = \frac{1}{9}$.

## 4.

Заданное условие можно трактовать следующим образом:

"С находится на расстоянии не более 9 см от А" - $C \in [0, 9]$\
"С находится на расстоянии не более 15 см от B" - $C \in [5, 20]$,

что сводится к условию $C \in [5, 9]$.

Вероятность случайного попадания точки в промежуток шириной 4 см на отрезке 20 см составляет $\frac{4}{20} = 0.2$.

## 5.

Каждая цифра телефонного номера не зависит от остальных. Предполагая, что на каждом месте номера может находиться каждая цифра и выбирая из 10 цифр для каждой позиции, получаем, что вероятность получить конкретный номер равна $(\frac{1}{10})^{7} = 1\cdot10^{-7}$.

## 6.

Т.к. порядок цифр в телефонном номере имеет значение, рассчитаем число размещений при выборе 2 элементов из 9 (цифры от 1 до 9).

$A^{2}_{9} = 9 \cdot 8 = 72$

Соответственно, необходимо будет перебрать 72 варианта, вероятность угадать с первого раза $\frac{1}{72}$.

## 7.

Если разрезать большой куб на 27 маленьких кубиков, то в формировании наружных граней большого куба будет участвовать только 26 кубиков, 1 окажется "запертым" в центре. Ни одна из граней такого кубика окрашена не будет. Если этот кубик попадет в любое другое место, кроме центра, при повторном формировании куба, то новый куб не будет белым.

Условие "черный кубик оказывается в середине" равносильно условию "вытягивать кубики с окрашенной гранью 26 раз при формировании граней большого куба". Число различных комбинаций, при которых можно вытащить 26 окрашенных кубиков - $A^{26}_{26} = 26!$. Общее число комбинаций, которыми можно вытащить 26 (не обязательно окрашенных!) кубиков - $A^{26}_{27} = 27!$. Тогда вероятность вытащить 26 окрашенных кубиков, не вытаскивая неокрашенный, равна $\frac{26!}{27!} = \frac{1}{27}$.

Для того, чтобы итоговый куб получился белым, нужно также, чтобы каждый из маленьких кубиков был повернут наружу своей окрашенной гранью. Вероятность этого для каждого кубика $\frac{1}{6}$. Т.к. у нас 26 "наружных" кубиков, каждый из которых располагается независимо от остальных, вероятность того, что все наружные кубики будут смотреть в нужном направлении $(\frac{1}{6})^{26} = \frac{1}{6^{26}}$.

Так как выбор кубиков и их расположение - независимые события, итоговая вероятность получить полностью белый куб равна произведению вероятностей A (вытащить все "правильные" кубики для граней) и B (вытащить все кубики в правильном положении).

$ P = \frac{1}{27} \cdot \frac{1}{6^{26}} = \frac{1}{27 \cdot 6^{26}}$
