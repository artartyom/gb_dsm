Косинус угла между этой прямой и осями координат напрямую зависит от расстояния исходных точек до оси Ox. Следовательно, в результате ортогонального преобразования получившаяся прямая не будет параллельна изначальной, что противоречит определению ортогонального преобразования.

### 3.2

```python
def plotCirclePolar(r):
    x = np.linspace(-r, r, 500)
    y = np.sqrt(r**2-x**2)
    plt.polar(x,y)
    plt.polar(x, -y)
    
plotCirclePolar(3)
```

### 3.3

```python
def plotLinePolar(k, b, lim1, lim2):
    x = np.linspace(lim1, lim2, 500)
    y = k*x + b
    plt.polar(x, y)
    
plotLinePolar(2,4, -1, 1)
```

## Задание 4

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 1000)
y1 = (np.exp(x)+x-1)/x
y2 = x**2 - 1

plt.plot(x, y1)
plt.plot(x, y2)
```

```python
def equations(p):
    x, y = p
    return (np.exp(x)+x*(1-y)-1, y-x**2+1)

x1, y1 = fsolve(equations, (4, 20))
x2, y2 = fsolve(equations, (2, 5))
x3, y3 = fsolve(equations, (-2, 2))
print("x1: {}, y1: {}".format(round(x1, 3), round(y1, 3)))
print("x2: {}, y2: {}".format(round(x2, 3), round(y2, 3)))
print("x3: {}, y3: {}".format(round(x3, 3), round(y3, 3)))
```
