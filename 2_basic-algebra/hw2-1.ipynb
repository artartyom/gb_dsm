## Задание 3

$\frac{x^2}{R^2}+\frac{y^2}{R^2}=1$

$x^2+y^2=R^2$

$y=\pm\sqrt{R^2-x^2}$

```python
def plotCircle(r):
    x = np.linspace(-r, r, 500)
    y = np.sqrt(r**2-x**2)
    plt.plot(x,y)
    plt.plot(x, -y)
    
plotCircle(3)
```

$\frac{x^2}{a^2}+\frac{y^2}{b^2}=1$

$\frac{y^2}{b^2}=1-\frac{x^2}{a^2}$

$y=\pm\sqrt{(1-\frac{x^2}{a^2})*b^2}$

```python
def plotEllipse(a, b):
    x = np.linspace(-a, a, 500)
    y = np.sqrt((1-(x/a)**2)*b**2)
    plt.plot(x, y)
    plt.plot(x, -y)
    
plotEllipse(1, 3)
```

$\frac{x^2}{a^2}-\frac{y^2}{b^2}=1$

$\frac{y^2}{b^2}=\frac{x^2}{a^2}-1$

$y=\pm\sqrt{(\frac{x^2}{a^2}-1)*b^2}$

```python
def plotHyperbola(a, b):
    x1 = np.linspace(-b, -a, 500)
    x2 = np.linspace(a, b, 500)
    y1Sq = ((x1**2/a**2)-1)*(b**2)
    y1 = np.sqrt(np.abs(y1Sq))*np.sign(y1Sq)
    y2Sq = ((x2**2/a**2)-1)*(b**2)
    y2 = np.sqrt(np.abs(y2Sq))*np.sign(y2Sq)    
    plt.plot(x1, y1)
    plt.plot(x1, -y1)
    plt.plot(x2, y2)
    plt.plot(x2, -y2)
    
plotHyperbola(1, 3)
```
